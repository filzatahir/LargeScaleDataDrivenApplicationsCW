```python
from pathlib import Path
CSV_Path = RAW / 'CarSales_Dataset.csv'

if CSV_Path.exists():
  print('CSV found at:', CSV_Path)
else:
  raise FileNotFoundError("CSV NOT FOUND!")
```

```python
import pandas as pd
df = pd.read_csv(CSV_Path)
print("Shape:", df.shape)
df.head(10)
```

```python
#remove duplicates
df_clean = df.drop_duplicates().copy()
print("original rows:", len(df))
print("after removing duplicates:", len(df_clean))
df_clean.head(10)
```

```python
#find all string (object) columns
str_cols = df_clean.select_dtypes(include='object').columns

#strip spaces from each
df_clean[str_cols] = df_clean[str_cols].apply(lambda col: col.str.strip())

print("trimmed space in:", len(str_cols), "columns")
df_clean.head(10)
```

```python
df_clean.dtypes
```

```python
#convert numeric columns to numbers
# ALL numeric columns
numeric_cols = [
    "Mileage",
    "Price",
    "Cost_of_Service",
    "Cost_of_Repair",
    "Latitude",
    "Longitude",
    "Year_of_Manufacturing",
    "Engine size"
]

for col in numeric_cols:
    if col in df_clean.columns:
        df_clean[col] = pd.to_numeric(df_clean[col], errors="coerce")

df_clean[numeric_cols].head()
```

```python
#clean up the date columns
date_cols = ["Date_of_Service", "Date_of_Accident"]

for col in date_cols:
    df_clean[col] = pd.to_datetime(
        df_clean[col],
        format="%d/%m/%Y",
        errors="coerce"
    ).dt.strftime("%Y-%m-%d")

df_clean[date_cols].head(20)
```

```python
#create the dealer table
dealer_cols = ["DealerName", "DealerCity", "Longitude", "Latitude"]

#extract unique dealer
dealer_df = (
    df_clean[dealer_cols]
    .drop_duplicates()
    .reset_index(drop=True)
)
#create DealerID (PK)
dealer_df.insert(0, "DealerID", range(1, len(dealer_df) + 1))

#save table
Dealer_File = PROC / "Dealer.csv"
dealer_df.to_csv(Dealer_File, index=False)

dealer_df.head()
dealer_df.shape
```

```python
#compare to original
print("Unique dealers in original df_clean :", df_clean['DealerName'].nunique())
print("Unique dealers in dealer_df        :", dealer_df['DealerName'].nunique())
```

```python
#merge DealerID to the original dataset
df_merged = df_clean.merge(
    dealer_df,
    on=["DealerName", "DealerCity", "Latitude", "Longitude"],
    how="left",
    validate="m:1"
)

df_merged.head()
```

```python
#build the car table
car_cols = [
    "CarID",
    "Manufacturer",
    "Model",
    "Engine size",
    "Fuel_Type",
    "Year_of_Manufacturing",
    "Mileage",
    "Price",
    "DealerID"
]

car_df = (
    df_merged[car_cols]
    .drop_duplicates()
    .reset_index(drop=True)
)

CAR_FILE = PROC / "Car.csv"
car_df.to_csv(CAR_FILE, index=False)

car_df.head()
car_df.shape
```

```python
#compare to original
print("Unique CarIDs in original df_clean:", df_clean["CarID"].nunique())
print("Unique CarIDs in car_df          :", car_df["CarID"].nunique())

car_df.head()
```

```python
#ensure dealer ids are unique + not regenerated
df_merged.groupby("DealerName")["DealerID"].nunique().sort_values(ascending=False).head(10)
```

```python
print("DealerIDs in dealer_df:", dealer_df["DealerID"].nunique())
print("DealerIDs in car_df   :", car_df["DealerID"].nunique())
```

```python
#create feature table
import pandas as pd
#extract unique features
feature_df = (
    df_merged[["Features"]]
    .dropna()                 #remove empty rows
    .drop_duplicates()        #keep only unique features
    .reset_index(drop=True)
)

#create FeatureID (Primary Key)
feature_df.insert(0, "FeatureID", range(1, len(feature_df) + 1))

#save feature table
FEATURE_FILE = PROC / "Feature.csv"
feature_df.to_csv(FEATURE_FILE, index=False)

feature_df.head(), feature_df.shape
```

```python
print("Unique features in original df_clean :", df_clean["Features"].nunique())
print("Unique features in feature_df       :", feature_df["Features"].nunique())

feature_df.head()
```

```python
#build carfeature table

#join df_merged with feature_df to get FeatureID for each row
carfeature_df = (
    df_merged[["CarID", "Features"]]
    .dropna(subset=["Features"])            #keep rows with features
    .merge(feature_df, on="Features", how="left", validate="m:1")
    [["CarID", "FeatureID"]]
    .drop_duplicates()
    .reset_index(drop=True)
)

#save join table
CarFeature_File = PROC / "CarFeature.csv"
carfeature_df.to_csv(CarFeature_File, index=False)

carfeature_df.head(), carfeature_df.shape
```

```python
#compare to original
print("Unique CarIDs in original:", df_clean["CarID"].nunique())
print("Unique CarIDs in CarFeature:", carfeature_df["CarID"].nunique())

print("\nUnique FeatureIDs in Feature table:", feature_df["FeatureID"].nunique())
print("Unique FeatureIDs in CarFeature:", carfeature_df["FeatureID"].nunique())

carfeature_df.head()
```

```python
service_cols = [
    "ServiceID",
    "CarID",
    "Date_of_Service",
    "ServiceType",
    "Cost_of_Service"
]

service_df = (
    df_merged[service_cols]
    .dropna(subset=["ServiceID"])        #only keep real service rows
    .drop_duplicates()
    .reset_index(drop=True)
)

Service_File = PROC / "Service.csv"
service_df.to_csv(Service_File, index=False)

service_df.head(), service_df.shape
```

```python
print("Unique ServiceIDs in original df_clean:", df_clean["ServiceID"].nunique())
print("Unique ServiceIDs in service_df      :", service_df["ServiceID"].nunique())

print("\nDo all ServiceIDs link to valid CarIDs?")
print(service_df["CarID"].isin(car_df["CarID"]).all())

print("\nMissing CarIDs in service table:", service_df["CarID"].isna().sum())
service_df.head()
```

```python
accident_cols = [
    "AccidentID",
    "CarID",
    "Date_of_Accident",
    "Description",
    "Cost_of_Repair",
    "Severity"
]

accident_df = (
    df_merged[accident_cols]
    .dropna(subset=["AccidentID"])         #keep ONLY real accident rows
    .drop_duplicates()
    .reset_index(drop=True)
)

Accident_File = PROC / "Accident.csv"
accident_df.to_csv(Accident_File, index=False)

accident_df.head(), accident_df.shape
```

```python
print("Unique AccidentIDs in original df_clean:", df_clean["AccidentID"].nunique())
print("Unique AccidentIDs in accident_df      :", accident_df["AccidentID"].nunique())

print("\nDo all AccidentIDs link to valid CarIDs?")
print(accident_df["CarID"].isin(car_df["CarID"]).all())

print("\nMissing CarIDs in accident table:", accident_df["CarID"].isna().sum())
accident_df.head()
```

```python
#final integrity check
print("1) CarIDs match:",
      df_clean["CarID"].nunique() == car_df["CarID"].nunique())

print("2) DealerIDs valid:",
      set(car_df["DealerID"]).issubset(set(dealer_df["DealerID"])))

print("3) Service + Accident CarIDs valid:",
      service_df["CarID"].isin(car_df["CarID"]).all() and
      accident_df["CarID"].isin(car_df["CarID"]).all())
```
